### 1. Compare my city with global data 

Suggestion : plot moving average rather than yearly average 

```python
Seoul = df[df['city']=='Seoul']
Berlin = df[df['city']=='Berlin']
```

```python
f,ax=plt.subplots(figsize=(12,5))
ax = sns.lineplot(x='year',y='10Year_moving_average',data=Seoul)
ax1 = sns.lineplot(x='year',y='10Year_moving_average',data=df1)
ax.legend(labels=['Seoul', 'World'])
plt.title("Seoul vs World")
plt.show()
```
